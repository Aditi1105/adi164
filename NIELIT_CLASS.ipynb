```python

import matplotlib.pyplot as plt
with open("/QueryResults.csv","r") as f:
  df=pd.read_csv(f,names=['DATE','TAG','POSTS'],header=0)
  
  df.DATE=pd.to_datetime(df.DATE)
  reshaped_df=df.pivot(index='DATE',columns='TAG',values='POSTS')
  # Replacing NaN values with zero
  reshaped_df.fillna(0,inplace=True)

  plt.figure(figsize=(16,10))  #change the dimension of plot by using figsize
 
 # xticks yticks-->datapoints into the graphs-->taken as a locator
  plt.xticks(fontsize=14)
  plt.yticks(fontsize=14)

  



  # x label and y label for labelling the graph
  plt.xlabel('Date',fontsize=18)
  plt.ylabel('Number of Posts',fontsize=18)
  plt.ylim(0,35000)  #limit of y-axis


  # we can use for-loop for generating other graphs also like c,c#,javascrpt
  for i in reshaped_df.columns:
    plt.plot(reshaped_df.index, reshaped_df[i],linewidth=3,label=reshaped_df[i].name)
    
    plt.legend(fontsize=16)   # for making a cloumn that wihich graph line shows for what








```
